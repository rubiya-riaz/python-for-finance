```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['MSFT', 'AAPL', 'F', 'GE']
```

```python
tickerDf = pd.DataFrame()
```

```python
for t in tickers:
    tickerDf[t]= wb.DataReader(t, data_source = 'yahoo', start = '1995-1-1')['Adj Close']    
```

```python
print(tickerDf.info)
```

```python
tickerDf.head()
```

# normalization to 100

```python
(tickerDf/ tickerDf.iloc[0]* 100).plot(figsize = (15, 6));
plt.show()
```

# Calculating the return of a portfolio of securities


```python
returns = (tickerDf/ tickerDf.shift(1)) - 1
returns.head()
```

```python
weights = ([0.30, 0.50, 0.10, 0.20])
```

```python
annual_returns = returns.mean()* 250
```

```python
annual_returns
```

```python
mat_mul = np.dot(annual_returns, weights)
```

```python
print(mat_mul)
```

```python
# round to 5 decimal places
```

```python
round = round(mat_mul, 5)
```

```python
print(round)
```

```python
percentage_returns = round * 100
```

```python
print(percentage_returns)
```
