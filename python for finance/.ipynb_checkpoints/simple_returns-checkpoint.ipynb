```python
import yfinance as yf
```

```python
# define ticker symbol
```

```python
tickerSymbol = 'AAPL'
```

```python
# get data on this ticker
```

```python
tickerData = yf.Ticker(tickerSymbol)
```

```python
# get the historical prices for this ticker
```

```python
tickerDf = tickerData.history(period = '1mo', start = '2015-1-1', end ='2020-4-30')
```

```python
# see your data
```

```python
# period= the frequency at which to gather the data; common options would include
```

```python
# '1d' (daily), '1mo'(monthly), '1y' (yearly). start = the date to start gathering the data
```

```python
# for example '2010-1-1'. end date= '2020-3-20'
```

```python
tickerDf
```

```python
close = tickerDf['Close']
```

```python
close
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# after shift number written is number of lags
```

```python
simple_return = (close /close.shift(1))-1
```

```python
print (simple_return)
```

```python
simple_return.plot(figsize=(8,5))
```

```python
# daily returns the d signifies
```

```python
avg_returns_d = simple_return.mean()
```

```python
# average returns the a signifies
```

```python
avg_returns_a=simple_return.mean()*250
```

```python
percentage_annual_average_returns = print(str(round(avg_returns_a, 5)*100) + '%')
```
