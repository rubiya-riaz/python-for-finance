```python
import numpy as np
import pandas as pd
import pandas_datareader as wb
import matplotlib.pyplot as plt
```

```python
# index names dont correspond with stock exchange names. for eg ixic is symbol for nasdaq. 
# nya  for nyse
```

```python
tickers = ['^GSPC','^IXIC','^GDAXI', '^DJI', '^NYA']
```

```python
ind_data=pd.DataFrame()
```

```python
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source = 'yahoo', start = '1997-1-2')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.dropna()
```

```python
# normalizing to 100
```

```python
normalized_data = ind_data/ ind_data.iloc[0]* 100
```

```python
print (normalized_data)
```

```python
import matplotlib.pyplot as plt
```

```python
normalized_data.plot(figsize=(15,6));
plt.show()
```

```python
# 1995 till 2000: dot com bubble therefore the peaks. NASDAQ: IXIC depicts best and is the highest peak because it consists
# primarily of tech companies. 
# nasdaq. highest in certain periods. shows performance connected to slightly sharper increases at times.
# germany GDAXI shows stability and growth
# 2007/2008 economic crisis and 2011 sovereign debt crisis

# simple returns
```

```python
# # dont use normalized returns here, use original dataset.
ind_returns = (ind_data-ind_data.shift(1)) - 1
```

```python
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean()* 250
```

```python
# gspc negative returns over last two decades. rest have positive returns over the last two decades.
# ixic nasdaq surpasses gspc sand p 500.
```

```python
annual_ind_returns/100
```
