```python
import yfinance as yf
```

```python
# define ticker symbol
```

```python
tickerSymbol = 'AAPL'
```

```python
# get data on this ticker
```

```python
tickerData = yf.Ticker(tickerSymbol)
```

```python
# get the historical prices for this ticker
```

```python
tickerDf = tickerData.history(period = '1mo', start = '2015-1-1', end ='2020-4-30')
```

```python
# see your data
```

```python
# period= the frequency at which to gather the data; common options would include
```

```python
# '1d' (daily), '1mo'(monthly), '1y' (yearly). start = the date to start gathering the data
```
